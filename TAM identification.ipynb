```python
def extract_5N(record_seq):

    upstream_5N = "TACTCGAGTTCATGTGCAGCTCCATAAGCAAAAGGGGATGATAAG"
    downstream_5N = "TCACATTCTTGCCCGCCTGA"

    upstream_index = record_seq.find(upstream_5N)
    downstream_index = record_seq.find(downstream_5N)
    
    result_seq = ""

    if upstream_index != -1 and downstream_index != -1:
        if upstream_index < downstream_index:
            result_seq = record_seq[upstream_index + len(upstream_5N):downstream_index]
        return result_seq
    else:
        return ""
```
